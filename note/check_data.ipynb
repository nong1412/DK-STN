```python
#去除模式数据的累加，再除以86400，得到W/s的值
def delete_add(data):
    #创建一个新数组，和data一样大小
    new_data=np.zeros(data.shape)
    new_data[0,:,:,:]=data[0,:,:,:]
    for i in range(1,len(data)):
        new_data[i,:,:,:]=data[i,:,:,:]-data[i-1,:,:,:]
    print(new_data[0:10,0,2,7])
    #new_data=np.true_divide(new_data,86400)
    return new_data
```
